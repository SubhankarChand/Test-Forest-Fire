# data cleaning

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

The dataset is convert into two sets based on Region from 122th index,we can make a new colun based on the region


1:"Bejaia Region Dataset"

2:"sidi-Bel Abbes Region Dataset"

```python
# devided the dataset into two part
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python

df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
# Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# rmove the 122th row 
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
# Fix spaces in coloumns names
df.columns = df.columns.str.strip()
df.columns
```

```python
## change the required columns as integer data types
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
## changing the other columns to float data type
objects=[features for features in df.columns if df[features].dtypes=='O' ]
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# Let ave the cleaning dataset
df.to_csv('Algerian_Foresr_Fires_cleaned_dataset.csv',index=False)
```

## Exploratory Data Analysis

```python
df_copy=df.drop(['day', 'month', 'year'],axis=1)
```

```python
df_copy.head()
```

```python

# Encoding the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python

# Apply Seaborn style
sns.set_theme(style="darkgrid")

# Plot histograms for all features
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
## precentage of pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## ploting pie chart
classlabels=["Fire", "Not Fire"]
plt.figure(figsize=(7,4))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("pie chart of classes")
plt.show()
```

```python
## corelation
df_copy.corr()
```

```python

# Select only the numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()

```

```python
## Box plot
sns.boxplot(df['FWI'],color='red')
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'not fire','fire')
```

```python
##Monthly Fire Analysis

# Filter the DataFrame for the specific region
dftemp = df.loc[df['Region'] == 1]

# Set up the figure size
plt.figure(figsize=(9, 4))

# Set the Seaborn style
sns.set_style('whitegrid')

# Plot the countplot
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel('Number of Fires', weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Algerian Forest Fire Analysis")
```
